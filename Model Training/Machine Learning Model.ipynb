```python
#importing libraries 

import numpy as np
import pandas as pd
#
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
```

```python
data = np.load('./data/data_pca_50_targets.npz')
data.files
```

```python
data.allow_pickle = True
```

```python
X = data['arr_0']
y = data['arr_1']
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test =train_test_split(X,y,test_size=0.20,stratify=y)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
model_SVC = SVC(probability=True)
param_grid = {'C':[0.5,1,10,20,30,50],
             'kernel':['rbf','poly'],
             'gamma':[0.1,0.001,0.05,0.01,0.002,0.005],
             'coef0':[0,1]}
```

```python
model_grid = GridSearchCV(model_SVC,param_grid=param_grid,
                          scoring='accuracy',cv=3,verbose=2)
```

```python
#Training the Data on the dataset

model_grid.fit(x_train,y_train)
```

```python
model_grid.best_params_
```

```python

model_final = model_grid.best_estimator_
model_final.get_params()
```

```python
#predicting the actual outputs
y_pred = model_final.predict(x_test)
y_pred
```

```python
##Classification Report

cr = metrics.classification_report(y_test,y_pred,output_dict=True)
pd.DataFrame(cr).T
```

```python
# Kappa Score

metrics.cohen_kappa_score(y_test,y_pred)
```

```python
## Area Under Curve

metrics.roc_auc_score(np.where(y_test=='male',1,0),
                      np.where(y_pred =='male',1,0))
```
