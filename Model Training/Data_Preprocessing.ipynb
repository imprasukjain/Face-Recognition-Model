```python
import numpy as np
import cv2
from glob import glob
import matplotlib.pyplot as plt
```

```python
fpath = glob('./data/female/*.jpg')
mpath = glob('./data/male/*.jpg')
```

```python
x=len(mpath)
y=len(fpath)
print(x,y)
```

```python
img = cv2.imread(fpath[-1])
img_RGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

haar = cv2.CascadeClassifier('./haar_model/haarcascade_frontalface_default.xml')
gray = cv2.cvtColor(img_RGB,cv2.COLOR_RGB2GRAY)
faces = haar.detectMultiScale(gray,1.5,5)
print(faces)
for x,y,w,h in faces:
    cv2.rectangle(img_RGB,(x,y),(x+h,y+w),(0,255,0),2)
    #Cropping the face
    roi = img_RGB[y:y+h,x:x+w]
    plt.imshow(roi)
    plt.axis('off')
    
```

```python
haar = cv2.CascadeClassifier('./haar_model/haarcascade_frontalface_default.xml')
```

```python
for i in range(len(fpath)):
    try:
        img = cv2.imread(fpath[i])
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        faces = haar.detectMultiScale(gray,1.5,5)
        for x,y,w,h in faces:
            roi = img[y:y+h,x:x+w]
            #Saving the image
            cv2.imwrite(f'./Cropped Data/female/female_{i}.jpg',roi)
            print("Image Transported Successfully")
    except:
        print("An Error Occured")
        
```

```python
for i in range(len(mpath)):
    try:
        img = cv2.imread(mpath[i])
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        faces = haar.detectMultiScale(gray,1.5,5)
        for x,y,w,h in faces:
            roi = img[y:y+h,x:x+w]
            #Saving the image
            cv2.imwrite(f'./Cropped Data/male/male_{i}.jpg',roi)
            print(f"{i+1} Image Transported Successfully")
    except:
        print("An Error Occured")
    finally:
        print("Process finished with exit code 0")
```
